```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
data = {
    'Student_ID': range(1, 11),
    'Math_Score': [78, 85, 92, np.nan, 88, 76, 95, 89, 101, 45], 
    'Science_Score': [80, 89, 94, 85, 78, np.nan, 88, 92, 99, 50],  
    'Attendance (%)': [95, 88, 92, 85, 90, 100, 75, np.nan, 98, 60],  
    'Study_Hours': [2, 3, 4, 1, 2.5, 3.5, 4.5, 5, 10, 0],  
    'Participation': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'No', 'No']
}

df = pd.DataFrame(data)
```

```python
df.isnull().sum()
```

```python
features = ['Math_Score','Science_Score','Attendance (%)']

for feature in features:
    df[feature].fillna(df[feature].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
# Detect outliers
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Math_Score', 'Science_Score', 'Study_Hours']])
plt.show()
```

```python
def remove_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR

    df[col] = np.where(df[col] > upper_bound, upper_bound, 
                          np.where(df[col] < lower_bound, lower_bound, df[col]))
```

```python
columns = ['Math_Score', 'Science_Score', 'Study_Hours']
for col in columns:
    remove_outliers(df,col)
```

```python
df
```

```python
df['Study_Hours'] = np.log1p(df['Study_Hours'])  # Log transformation
```

```python
df
```

```python
# Handling missing values (Fill NaN with column mean)
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df
```

```python
# Exclude categorical column and Student_ID
numerical_cols = ['Math_Score', 'Science_Score', 'Attendance (%)', 'Study_Hours']
```

```python
df
```

```python
# Min-Max Scaling
min_max_scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax[numerical_cols] = min_max_scaler.fit_transform(df[numerical_cols])
```

```python
df
```

```python
# Z-score Standardization
zscore_scaler = StandardScaler()
df_zscore = df.copy()
df_zscore[numerical_cols] = zscore_scaler.fit_transform(df[numerical_cols])
```

```python
df
```

```python
# Display results
print("Original Data:")
print(df)
print("\nMin-Max Scaled Data:")
print(df_minmax)
print("\nZ-score Standardized Data:")
print(df_zscore)
```
