```python
import numpy as np
import pandas as pd
import seaborn as sbn
```

```python
df=pd.read_csv('student.csv')
```

```python
df
```

```python
df.describe()
```

```python
q1=df.math_score.quantile(0.25)
q3=df.math_score.quantile(0.75)

q1,q3
```

```python
IQR=q3-q1
IQR
```

```python
LL=q1-1.5*IQR
UL=q3-1.5*IQR
LL,UL
```

```python
df[(df.math_score<LL)|(df.math_score<UL)]
```

```python
df[(df.math_score<LL)&(df.math_score<UL)]
```

```python
df['zscore']=(df.math_score-df.math_score.mean())/df.math_score.std()
df
```

```python
sbn.boxplot(df['math_score'])
```

```python
mean_value = int(df['math_score'].mean())
df.loc[(df['math_score'] < LL) | (df['math_score'] > UL), 'math_score'] = mean_value
```

```python
sbn.boxplot(df['math_score'])
```
